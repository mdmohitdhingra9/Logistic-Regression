```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
titanic_data = pd.read_csv('train.csv')
```

```python
titanic_data.head(2)
```

```python
titanic_data.info()
```

```python
titanic_data.isnull()
```

```python
titanic_data.dropna(inplace=True)
```

```python
# get dummies to convert text to values
sex = pd.get_dummies(titanic_data['Sex'], drop_first=True)
```

```python
embark =  pd.get_dummies(titanic_data['Embarked'], drop_first=True)
```

```python
pclass =  pd.get_dummies(titanic_data['Pclass'], drop_first=True)
```

```python
# concat extra columns in actual data
titanic_data = pd.concat([titanic_data,sex,embark,pclass], axis=1)
```

```python
titanic_data
```

```python
titanic_data.drop(['Name','Sex','PassengerId','Ticket'], axis=1, inplace=True)
```

```python
titanic_data.drop(['Cabin'], axis=1, inplace=True)
```

```python
titanic_data.head(5)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
X = titanic_data.drop('Survived', axis=1)
Y = titanic_data['Survived']
```

```python
x_train, x_test, y_train, y_test=train_test_split(X, Y, test_size=0.3, random_state=41)
```

```python
logReg = LogisticRegression()
logReg.fit(x_train, y_train)
```

```python
# prediction
predictions = logReg.predict(x_test)
```

```python
predictions
```

```python
#use classification report for precision
from sklearn.metrics import classification_report
classification_report(y_test, predictions)
```

```python
#for accuracy of model use accuracy score
from sklearn.metrics import accuracy_score
accuracy_score(y_test, predictions)
```

```python
# use confustion matrix to get the above accuracy
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, predictions)
# confusion matrix = sum of (a[0,0]+a[1,1]) / (a[0,0]+a[0,1]+a[1,0]+a[1,1]) = 41/54 = 0.74
```
